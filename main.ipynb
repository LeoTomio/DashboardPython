```python
# Passo a Passo
# Passo 1: Importar a base de dados
# Passo 2: Visualizar a base de dados
# Passo 3: Corrigir as cagadas da base de dados
# Passo 4: Primeira analise do cancelamento dos clientes ( Qual a % de clientes que cancelou)
# Passo 5: Analise a causa do cancelamento dos clientes
```

```python
# Passo 1:
import pandas as pd

tabela = pd.read_csv("cancelamentos_sample.csv")
tabela = tabela.drop(columns="CustomerID")

# Passo 2:
display(tabela)

# Passo 3: Corrigir as cagadas da base de dados
tabela = tabela.dropna()
display(tabela.info())
```

```python
# Passo 4: Primeira analise do cancelamento dos clientes ( Qual a % de clientes que cancelou)



display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True))
# exibir em porcentagem ja : display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["duracao_contrato"].value_counts())
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
agrupamento = tabela.groupby("duracao_contrato").mean(numeric_only=True)

display(agrupamento)
```

```python
# Todos os clientes do contrato mensal cancelaram
```
